### 2.b Introduction to Pandas - Creating loss triangles and basic chainladder method

```python
df['Accident Year'] = [int(x[-4:]) for x in df['Accident Date']]
df['Reporting Year'] = [int(x[-4:]) for x in df['Report Date']]
df['Dev'] = df['Reporting Year'] - df['Accident Year'] + 1
df.head()
```

```python
reserve_data = df['Total Claim Amount'].groupby([df['Accident Year'],df['Dev']]).sum().reset_index()
```

```python
reserve_data
```

```python
reserve_data['cumsum'] = reserve_data['Total Claim Amount'].groupby(reserve_data['Accident Year']).cumsum()
reserve_data
```

```python
cumu_tri = pd.pivot_table(reserve_data, index = ["Accident Year"], columns = ["Dev"], values = ["cumsum"]).round(2)
cumu_tri
```

```python
def GetLdf(tri,dp=3):
    index = len(tri.columns)
    LDF = []
    for c in range(1, index):
        ldf = tri.iloc[0:(index - c), c].sum() / tri.iloc[0:(index - c), c - 1].sum()
        LDF.append(ldf)
    round_ldf = [round(elem, dp) for elem in LDF]
    return round_ldf

def GetUlt(tri, f = None, tailf = 1, showult = True):
    if f is None:
        f = GetLdf(tri,dp=5)
    f.append(tailf)
    new_tri = tri.copy()
    n = len(new_tri.columns)
    new_tri['Ult'] = 1
    for k in range(1,n + 1):
        new_tri.iloc[n-k:n,k] = new_tri.iloc[n-k:n,k - 1]*f[k-1]
    return round(new_tri)

def PlotTri(tri):    
    for i in range(0,len(tri.columns)):
        tri.iloc[i,:].plot(figsize=(12,10),marker = '${}$'.format(i+1),ms= 10)
    plt.show()
```

```python
ldf = GetLdf(cumu_tri)
ldf
```

```python
GetUlt(cumu_tri)
```

```python
PlotTri(cumu_tri)
```
